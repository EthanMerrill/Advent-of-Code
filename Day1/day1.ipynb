```python
import pandas as pd

f = open ('elves.txt', 'r')
data = f.read()
f.close()
```

```python
data = data.splitlines()

df = pd.DataFrame(data, columns = ['data'])
df.insert(1, 'elfNumber', 0)
# about to write some ugly inefficient code

```

```python

elfNumber = 1
for index, row in df.iterrows():
    if len(row['data']) == 0:
        elfNumber = elfNumber + 1
        print("TESTOING", elfNumber)
    else:
        df.loc[index, 'elfNumber']= elfNumber
```

```python
df.drop(df[df['data'] == ''].index, inplace = True)
print(df)
df['data'] = df.data.astype(int)
dfmod = df.groupby(by = ['elfNumber']).sum(numeric_only=True)
dfmod.data.max()
```
